## Exercise 1
A chart is basically made of: *visual encoding* and *chart apparatus*. Chart apparatus refers to features related to the type of chart that has been chosen, while the visual encoding elements are customized. In this case, the chart shows a combination of different graphs:
* one line chart in the upper left, showing the children poisonings reported between years 1983, 2018
* one bar chart (vertical) in the lower left, showing the ten most involved substances in child poisonings
* one chord diagram on the right,
First, we need to make a first distinction in the visual encoding elements, between *marks* and *annotations* (or attributes). Marks consist in the way the data is reported in the chart e.g. dots, lines, area and so on. Marks are also related to the dimension we deal with. Annotations instead could be quantity, size, colours and every other attribute that can represent a quantity or a relation.
The main goal when representing data is to find the right blend of marks and annotations that most effectively portray the angle of analysis you wish to show.

In this case, the marks present are *lines* representing:
* the reported poisonings from 1983 to 2018, with quantity represented with the position
* the 10 most involved substances, with quantity represented with the size
* the relation between age of the children and reasons of the poisonings
As mentioned before, annotations could be quantitative or categorical attributes. In this case, quantitative attributes present in this graph are:
* position of the line respect to the Y axis in the line chart
* size (in length) of the line in the bar plot
* size (width) of the line connecting children's age and reasons.
While the graphs on the left use only one color, the graph on the right uses different tonalities of blue to distinguish the various categories that are connected through a line in order to represent the correlation between a certain age and the reason of the poisoning.
Finally, in the upper right part of the chart, we can notice 3 numbers. These are annotations, giving us some information highlighting what the chord diagram is telling us.

## Exercise 2
The geometric projection in map representation refers to represent the surface of the Earth, that is curve, on a flat map. The first thing to consider is that **every projection is distorted**, and typically the larger the area is, the grater the distortion is. We need to choose carefully the type of map to use since no projection can accommodate all different cases.
While choosing, we can consider some factors: like scope of view, distance from the equator, our focus on the map and so on.
Different map projections are available:
* **Mercator**: is the most diffused map. It has a large distortion far from the equator. Good for focusing on a specific region, not so good for global view
* **Albers**: also, this map should not be used for global view, but is pretty good at country level.
* **Lambert**: since it is a spherical projection, it is good for hemisphere or country-level view, as the european environment agency confirms
* **Winkel-Triple**: this is an excellent map for global view, as national geographic confirms
* **Mollweide**: accurate ocean areas, this map is more adequate for atmospheric mappings and flight paths.

As you can notice in this figure, Greenland is represented much bigger respect to the reality in mercator map projection. Let's assume we want to show a certain quantity of area. In a global view, the Winkel-Tripel projection results in a more accurate visualization respect to the Mercator one (that could be misleading!).

## Exercise 3
UMAP (*Uniform Manifold Approximation and Projection*) is a dimensionality reduction technique used for visualizing high-dimensional data. The main goal is to preserve local and global structure of the data in a lower dimensional space.
Parameters taken by the UMAP function, that defines its behaviour, are: number of neighbors, minimum distance, number of components and metric.

### Number of neighbors - n_neighbors
This parameter defines the number of neighbors that is used to install the connections between each point and his respective neighbors. The creation of a connected graph is needed to approximate the topology. High values of number of neighbors result in a more global view of the data. Low values of number of neighbors instead result in a more local structure. Typically, values between 5 and 50 are suggested.

### Minimum distance - min_dist
In the low dimensional embedding, this parameter defines the minimum distance between the points. Smaller values result in better preserving of local structures, while higher values result in a more preserved global structure.

### Number of components - n_components
It defines the desired number of dimensions in the low dimensional embedding space. By default, it is 2-dim, but if our data has an intrinsic dimensionality greater than 2, we can provide any positive integer number.

### Metric
This parameter determines the distance metric to measure the distance between data points in the initial space. The choice of the metric depends on the data and the problem. Typically Euclidean space is used.

### Random state
This parameter refers to the random number generator used in UMAP while initializing. Using a fixed integer, it is possible to reproduce the same result

As you can see at this [link]('https://pair-code.github.io/understanding-umap/'), varying number of neighbors and minimum distance lead to very different results.

```python
from IPython.display import Image

Image(filename='umap.png')
```

In the figure above, UMAP is applied to a 10-dim dataset containing 300 points with different values for *n_neighbors* and *min_dist*. As mentioned before, lower values of number of neighbors and minimum distance better preserve local structures, while higher values better preserve the global structure of the data.
