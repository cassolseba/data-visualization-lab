## Ex. 4 â€” Produce some plots from the renewable energy consumption dataset

```python
import numpy as np
import pandas as pd

rec = pd.read_csv("renewable-energy-consumption.csv")

rec_no_na = rec.dropna()

rec = rec[rec['Year'] == 2016]

years = rec_no_na['Year'].unique().tolist()

rec_no_na
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

sns.stripplot(y=rec_no_na['Traditional biofuels (terra-watt-hours)'], x=rec_no_na['Year'], data=rec_no_na)
plt.xticks(rotation = 90)
plt.xticks(ticks=np.arange(0, len(rec_no_na['Year']), 5), labels=[str(years[i]) for i in range(0, len(years), 5)])
plt.title("Traditional biofuels consumption in the world from 1965 to 2016")
plt.ylim(0, max(rec_no_na['Traditional biofuels (terra-watt-hours)']) + 1000)


# labels = [str(year_l[i]) for i in range(0,len(year_l), 5)]
plt.show()
```

Another visualization for the same data could be a bar plot!

```python
fig = plt.figure()

rec_every_5_years = rec_no_na[rec_no_na['Year'] % 5 == 0]
five_years = rec_every_5_years['Year'].unique().tolist()
sns.barplot(
    data = rec_no_na,
    y = rec_every_5_years['Hydropower (TWh)'],
    x = rec_every_5_years['Year'],
    color='tab:blue',
)

plt.xticks(
    rotation = 60,
    ticks=np.arange(0, len(rec_every_5_years['Year']), 2),
    labels=[str(five_years[i]) for i in range(0, len(five_years), 2)]
)

plt.title("Increasing of hydropower energy consumption (in tera-watt) from 195 to 2015", fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 10), dpi = 100)

plt.plot('Year', 'Traditional biofuels (terra-watt-hours)', data=rec_no_na, linestyle="--", color="blue")
plt.plot('Year', 'Wind (Terawatt-hours)', data=rec_no_na, linestyle="-", color="red")
plt.plot('Year', 'Other renewables (modern biofuels, geothermal, wave & tidal) (terawatt-hours)', data=rec_no_na, linestyle="-", color="green")
plt.plot('Year', 'Solar PV (Terawatt-hours)', data=rec_no_na, linestyle="-", color="yellow")
plt.plot('Year', 'Hydropower (TWh)', data=rec_no_na, linestyle="-", color="grey")

plt.xticks(
    fontsize = 15,
    rotation = 30,
    ha = 'center',
    va = 'top'
)

plt.tick_params(
    axis='both',
    which='both',
    color='red',
    width=2,
    length=8,
    direction='inout',
    bottom=True,
    left=True,
    grid_alpha=0
)

plt.ylim(0, 14700)

plt.legend(fontsize = 14)
plt.title("Consumption of energy in tera-watt in the world", fontsize=25)

plt.show()
```

Another possible good visualization could be the amount of renewable energy consumed in one year, shown by a pie plot.

```python
rec_2016 = rec_no_na[rec_no_na['Year'] == 2016]

rec_2016
```

```python
# pick the values for year 2016
consumes_2016 = []
for column in rec_2016.columns:
    consumes_2016.append(rec_2016[column].to_list()[0])
consumes_2016 = consumes_2016[3:]

# calculate total consumption in tera-watt
total = 0
for i in range(len(consumes_2016)):
    total += consumes_2016[i]

# compute percentages for pie plot
percentages = []
for j in range(len(consumes_2016)):
    percentages.append(round((consumes_2016[j]*100)/total))

# get column names
columns = rec_2016.columns.to_list()[3:]
```

```python
fig = plt.figure(figsize=(8, 8), dpi = 300)

plt.pie(consumes_2016, labels=columns, autopct='%1d%%')

plt.title("Energy consumption in 2016")
```
