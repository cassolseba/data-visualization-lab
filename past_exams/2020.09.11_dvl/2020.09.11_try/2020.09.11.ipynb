## Exercise 5
Using `madelon.csv`, use PCA (2d), coloring the samples according to their class. Then use UMAP and compare the results.

```python
# import
import pandas as pd
```

In this case, for this csv file, no name columns are provided and the separator is a space. Thus, we need to define in `read_csv` function both the separator and the header.

```python
# read data
madelon = pd.read_csv('../madelon.csv', sep=' ', header=None)
```

```python
# view madelon dataset
madelon.head()
```

```python
# instantiate PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
```

```python
# preapare the dataset for PCA
feat_cols = madelon.columns[:-1]

feat_data = madelon[feat_cols]

feat_data.head()
```

```python
# compute PCA
pca_result = pca.fit_transform(feat_data)

# save to dataframe
madelon['pca-one'] = pca_result[:,0]
madelon['pca-two'] = pca_result[:,1]

print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
# import
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# view
plt.figure(figsize=(16,10))

sns.scatterplot(
    x=madelon['pca-one'],
    y=madelon['pca-two'],
    hue=500,
    palette=sns.color_palette("hls", 2),
    data=madelon,
    legend="full",
    alpha=0.3
)

plt.show()
```

Here is the result, not a good separation. Let's try with UMAP now.

```python
# import UMAP
import umap.umap_ as umap
```

```python
# compute UMAP
umap_result = umap.UMAP(n_neighbors=5, random_state=42).fit_transform(feat_data)
```

```python
# plot with matplotlib
plt.scatter(umap_result[:, 0], umap_result[:, 1], s=5, c=madelon[500], cmap='Spectral')
plt.title('UMAP projection of the Madelon dataset', fontsize=24);
```

```python
# import results in madelon dataset
madelon['umap-one'] = umap_result[:,0]
madelon['umap-two'] = umap_result[:,1]
```

```python
# plot with seaborn
plt.figure(figsize=(16,10))

sns.scatterplot(
    x=umap_result[:, 0],
    y=umap_result[:, 1],
    hue=500,
    palette=sns.color_palette("hls", 2),
    data=madelon,
    legend="full",
    alpha=0.3
)

plt.show()
```

Let's try to plot a single plot with all the results!

```python
plt.figure(figsize=(25, 10))

plt.suptitle("PCA vs. UMAP", fontsize=20)

plt.subplot(1, 2, 1)
plt.scatter(madelon['pca-one'], madelon['pca-two'], s=5, c=madelon[500], cmap='Spectral')
plt.title("PCA")

plt.subplot(1, 2, 2)
plt.scatter(madelon['umap-one'], madelon['umap-two'], s=5, c=madelon[500], cmap='Spectral')
plt.title("UMAP")

plt.tight_layout()

plt.show()
```

```python
# now with seaborn

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 10), dpi=200)

plt.suptitle("Madelon: pca vs. umap")

sns.scatterplot(
    x=madelon['pca-one'],
    y=madelon['pca-two'],
    hue=500,
    palette=sns.color_palette("hls", 2),
    data=madelon,
    legend="full",
    alpha=0.3,
    ax=ax1
)
ax1.set_title("PCA")

sns.scatterplot(
    x=madelon['umap-one'],
    y=madelon['umap-two'],
    hue=500,
    palette=sns.color_palette("hls", 2),
    data=madelon,
    legend="full",
    alpha=0.3,
    ax=ax2
)
ax2.set_title("UMAP")

plt.tight_layout()

plt.show()
```

## Exercise 6
Replicate the Sankey Plot using `sankey.csv`.

```python
# import
import pandas as pd
```

```python
# load sankey data
sankey = pd.read_csv("../sankey.csv")
```

```python
# view
sankey.head()
```

```python
countries = sankey['FROM'].unique().tolist()

countries_to = sankey['TO'].unique().tolist()

for country in countries_to:
    if country not in countries:
        countries.append(country)

countries
```

```python
indexes = []
for i in range(len(countries)):
    indexes.append(i)

indexes
```

```python
# replace name of countries with indexes
for i in range(len(sankey['FROM'])):
    sankey['FROM'][i] = countries.index(sankey['FROM'][i])

sankey['FROM']
```

```python
# replace name of countries with indexes
for i in range(len(sankey['TO'])):
    sankey['TO'][i] = countries.index(sankey['TO'][i])

sankey['TO']
```

```python
# set colors
colors = ["blue", "red", "orange", "orange", "blue", "green", "green", "red", "yellow", "purple", "purple", "blue", "yellow", "blue", "green", "green"]
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad=15,
        thickness=20,
        line = dict(color = colors, width = 0.5),
        label = countries,
        color=colors
    ),
    link = dict(
        source = sankey['FROM'],
        target = sankey['TO'],
        value = sankey['VALUE']
    ))])

fig.update_layout(title_text="Sankey Diagram", font_size=10)
fig.show()
```
