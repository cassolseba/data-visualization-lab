# Ex. 4 | Find a data visualization

```python
import matplotlib.pyplot as plt
import pandas as pd

forest_fires = pd.read_csv("forestfires.csv")

forest_fires
```

```python
len(forest_fires['month'].unique()) # how many months are in the dataset --> entire year

# months = forest_fires['month'].unique().tolist() # get list of months
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'] # ordered months
# compute amounts of burned area for every month
burned_area = []
for month in months:
    forest_fires_ = forest_fires[forest_fires['month'] == month]
    burned_area.append(forest_fires_['area'].sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

palette = sns.color_palette('deep') # pick color palette

plt.figure(figsize=(10, 8))
plt.bar(list(range(1, 13)), burned_area, width=1, color=palette)
plt.xticks(
    ticks=list(range(1, 13)), # tick positions
    labels=months, # tick labels
    rotation=0, # rotation of labels
    fontsize=12, # labels fontsize
    horizontalalignment='center', # horizontal alignment
    alpha=.7 # transparency
)
plt.ylabel('Burned area of the forest (in ha)')
plt.xlabel('Month')
plt.title('Amount of forest burned in Montesinho park in 2007')
plt.show()
```

# Ex. 5 | Pt.1 --> Heatmap

```python
import pandas as pd

forest_fires = pd.read_csv("forestfires.csv")

forest_fires
```

```python
jan_ff = forest_fires[forest_fires['month'] == 'jan'] # select by month 'jun'

jul_ff = forest_fires[forest_fires['month'] == 'jul'] # select by month 'jul'

jul_ff
```

```python
# prepare dataset for July
jul_temp = jul_ff.iloc[:, 0:2] # copy x and y columns (pick first two columns)
jul_temp['temp'] = jul_ff['temp'] # copy temp column
jul_temp.reset_index(inplace=True) # reset indexes
jul_temp = jul_temp.iloc[:, 1:] # remove old indexes

jan_temp = jan_ff.iloc[:, 0:2] # copy x and y columns
jan_temp['temp'] = jan_ff['temp'] # copy temp column
jan_temp.reset_index(inplace=True) # reset indexes
jan_temp = jan_temp.iloc[:, 1:] # remove old indexes
```

```python
import numpy as np

mx_jul = np.zeros((9,9))
for i in range(len(jul_temp)):
    row = jul_temp.iloc[i] # pick the i-rows of the dataset
    mx_jul[int(row[0])-1, int(row[1])-1] = row[2] # set in matrix at x (row[0]-1) and y (row[1]-1) value temp (row[2])

print(mx_jul)

mx_jan = np.zeros((9,9))
for i in range(len(jan_temp)):
    row = jan_temp.iloc[i] # pick the i-rows of the dataset
    mx_jan[int(row[0])-1, int(row[1])-1] = row[2] # set in matrix at x (row[0]-1) and y (row[1]-1) value temp (row[2])

print(mx_jan)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(25, 10))

plt.suptitle('Heatmap of temperature in Monteshino park', size = 25)

sns.heatmap(ax = axes[0], data = mx_jul, cmap=sns.light_palette('orange', as_cmap=True))
axes[0].set_title('Temperature in July', size = 20)

sns.heatmap(mx_jan, ax = axes[1], cmap=sns.light_palette('orange', as_cmap=True))
axes[1].set_title('Temperature in January', size = 20)

#plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()
```

# Ex.5 | Pt.2 --> UMAP

```python
import pandas as pd

forest_fires = pd.read_csv("forestfires.csv")

forest_fires
```

```python
months = forest_fires['month'].unique()
months
```

```python
# assign the correct number to months
num_month = {
    'mar': 3,
    'oct': 10,
    'aug': 8,
    'sep': 9,
    'apr': 4,
    'jun': 6,
    'jul': 7,
    'feb': 2,
    'jan': 1,
    'dec': 12,
    'may': 5,
    'nov': 11
}

num = [] # get number of months
for month in forest_fires['month']:
    num.append(num_month[month])

print(num)
```

```python
forest_fires = forest_fires.iloc[:, 4:8] # pick only the columns we need: FFMC, DMC, DC, ISI
```

```python
forest_fires['month'] = num
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import umap.umap_ as umap
```

```python
trans = umap.UMAP(n_neighbors=5, random_state=42).fit(forest_fires.iloc[:, :4]) # exclude number of months
```

```python
plt.figure(figsize= (15, 8))
plt.scatter(trans.embedding_[:, 0], trans.embedding_[:, 1], s=5, c=forest_fires['month'], cmap='Spectral')
plt.title('UMAP', fontsize=24)
```

# Ex. 6 | Recreate that plot

```python
import pandas as pd

cars = pd.DataFrame(
    {'company': ['Ford', 'GM', 'Toyota'],
     2011 : [1, 2, 3],
     2012 : [4, 2, 8],
     2013 : [6, 7, 5],
     2014 : [8, 10, 10],
     2015 : [9, 12, 6]}
)

cars
```

```python
cars.set_index('company', inplace=True) # set the column's company as index

cars.loc[['Ford']] # test for getting Ford row
```

```python
import seaborn as sns
sns.set_style('white')
plt.stackplot(
    [1, 2, 3, 4, 5],
    cars.loc[['Ford']],
    cars.loc[['GM']],
    cars.loc[['Toyota']],
    labels=('Ford', 'GM', 'Toyota')
)
plt.xticks([1, 2, 3, 4, 5], ['2011', '2012', '2013', '2014', '2015'])
plt.title(
    'Car Sales in US', # title
    color='orange', # color
    loc='left' # position
)
plt.xlabel('Year')
plt.ylabel('Number of cars (M)')
plt.legend(loc = "upper left") # legend in upper left position

plt.show()
```
